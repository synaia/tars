```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.memory import ChatMessageHistory

from rich import print
from dotenv import dotenv_values

secret = dotenv_values('../../.secret')

chat = ChatOpenAI(model="gpt-3.5-turbo-1106", temperature=0.2, openai_api_key=secret['OPEN_AI_API_KEY'])

import langchain
langchain.debug = True

chat_history = ChatMessageHistory()

print(chat)
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant. Answer all questions to the best of your ability. Use a short response.",
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)

chain = prompt | chat

# print(chain)
```

```python
# chat_history.add_ai_message(None)
chat_history.add_user_message("Michael has 20 apples, he shared 5 apples with Sofia, how many apples does Michael keep?")
response = chain.invoke({"messages": chat_history.messages})

print(response)
```

```python
chat_history.add_ai_message(response.content)
chat_history.add_user_message("Sofia share 2 apples with her daughter, how many apples does Sofia keep?")
response = chain.invoke({"messages": chat_history.messages})
```

```python
chat_history.add_ai_message(response.content)
chat_history.add_user_message("After that, Michael gives 1 apple to Sofia's daughter, how many apples does Michael have left?")
response = chain.invoke({"messages": chat_history.messages})
```
